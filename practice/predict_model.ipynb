```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import load_model
```

```python
path = 'C:/Users/kjh97/prac/trainset2/result_dataset.npy'
train_x, train_y, valid_x, valid_y, test_x, test_y = np.load(path, allow_pickle = True)
```

```python
train_x.shape, train_y.shape, valid_x.shape, valid_y.shape, test_x.shape, test_y.shape
```

```python
model = load_model('jong_s pill_s99.h5')
```

```python
model.fit(train_x, train_y, epochs = 10)
```

```python
val_loss, val_acc = model.evaluate( valid_x, valid_y, verbose =2)
```

```python
model.summary()
```

```python
predictions = model.predict(test_x)
```
