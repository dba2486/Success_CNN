#### 데이터셋 라벨링을 위한 함수

```python
def parse_data(folder, labels):
    path_train1 = glob.glob(folder+'/Carbamolpro/*.png')
    path_train2 = glob.glob(folder+'/Clodafen/*.png')
    path_train3 = glob.glob(folder+'/Gensokpeu/*.png')
    path_train4 = glob.glob(folder+'/Puligen/*.png')
    path_train5 = glob.glob(folder+'/Soxazone/*.png')
    path_train6 = glob.glob(folder+'/Furina/*.png')
    path_train7 = glob.glob(folder+'/Relaxel/*.png')
    list_x=[]
    list_y=[]
    for i in range(len(path_train1)):    
        path_img1 = path_train1[i]
        #image1 = Image.open(path_img1)
        image1 = cv2.imread(path_img1)
        arr = np.array(image1)
        list_x.append(arr)
        
        start1 = path_img1.index('set/')
        end1 = path_img1.index('\\')
        label1 = path_img1[start1+4:end1]
        label_index1 = labels.index(label1)
        list_y.append(label_index1)

    for i in range(len(path_train2)):  
        path_img2 = path_train2[i]
        #image2 = Image.open(path_img2)
        image2 = cv2.imread(path_img2)
        arr = np.array(image2)
        list_x.append(arr)
        
        start2 = path_img2.index('set/')
        end2 = path_img2.index('\\')
        label2 = path_img2[start2+4:end2]
        label_index2 = labels.index(label2)
        list_y.append(label_index2)
        
    for i in range(len(path_train3)): 
        path_img3 = path_train3[i]
        #image3 = Image.open(path_img3)
        image3 = cv2.imread(path_img3)
        arr = np.array(image3)
        list_x.append(arr)
        
        start3 = path_img3.index('set/')
        end3 = path_img3.index('\\')
        label3 = path_img3[start3+4:end3]
        label_index3 = labels.index(label3)
        list_y.append(label_index3)
        
    for i in range(len(path_train4)): 
        path_img4 = path_train4[i]
        #image4 = Image.open(path_img4)
        image4 = cv2.imread(path_img4)
        arr = np.array(image4)
        list_x.append(arr)
        
        start4 = path_img4.index('set/')
        end4 = path_img4.index('\\')
        label4 = path_img4[start4+4:end4]
        label_index4 = labels.index(label4)
        list_y.append(label_index4)
        
    for i in range(len(path_train5)):
        path_img5 = path_train5[i]
        #image5 = Image.open(path_img5)
        image5 = cv2.imread(path_img5)
        arr = np.array(image5)
        list_x.append(arr)
        
        start5 = path_img5.index('set/')
        end5 = path_img5.index('\\')
        label5 = path_img5[start5+4:end5]
        label_index5 = labels.index(label5)
        list_y.append(label_index5)
        
    for i in range(len(path_train6)):
        path_img6 = path_train6[i]
        #image6 = Image.open(path_img6)
        image6 = cv2.imread(path_img6)
        arr = np.array(image6)
        list_x.append(arr)
        
        start6 = path_img6.index('set/')
        end6 = path_img6.index('\\')
        label6 = path_img6[start6+4:end6]
        label_index6 = labels.index(label6)
        list_y.append(label_index6)
        
    for i in range(len(path_train7)):
        path_img7 = path_train7[i]
        #image7 = Image.open(path_img7)
        image7 = cv2.imread(path_img7)
        arr = np.array(image7)
        list_x.append(arr)
        
        start7 = path_img7.index('set/')
        end7 = path_img7.index('\\')
        label7 = path_img7[start7+4:end7]
        label_index7 = labels.index(label7)
        list_y.append(label_index7)
        
    return np.array(list_x), np.array(list_y)
```

#### 학습모델 제작

```python
X = tf.keras.layers.Input(shape=[37, 68, 3])
print('X', X)
#conv1-1, 1-2
H = tf.keras.layers.Conv2D(8, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(X)
print('H', H)
H = tf.keras.layers.Conv2D(8, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(X)
print('H', H)
#maxpool1
H = tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2)(H)
print('H', H)
#conv2-1, 2-2
H = tf.keras.layers.Conv2D(16, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(H)
print('H', H)
H = tf.keras.layers.Conv2D(16, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(H)
print('H', H)
#maxpool2
H = tf.keras.layers.MaxPool2D( pool_size = 2, strides = 2)(H)
print('H', H)
#conv3-1, 3-2, 3-3
H = tf.keras.layers.Conv2D(32, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(H)
print('H', H)
H = tf.keras.layers.Conv2D(32, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(H)
print('H', H)
#maxpool3
H = tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2)(H)
print('H', H)
#conv4-1, 4-2, 4-3
H = tf.keras.layers.Conv2D(64, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(H)
print('H', H)
H = tf.keras.layers.Conv2D(64, kernel_size = 3, strides = 1, padding = 'same', activation = 'relu')(H)
print('H', H)
#maxpool4
H = tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2)(H)
print('H', H)
#conv5-1, 5-2, 5-3
#maxpool5
#flatten
H = tf.keras.layers.Flatten()(H)
print('H', H)
#dense
H = tf.keras.layers.Dense(128, activation = 'relu')(H)
print('H', H)
H = tf.keras.layers.Dense(32, activation = 'relu')(H)
print('H', H)
Y = tf.keras.layers.Dense(7, activation = 'softmax')(H)
print('Y', Y)
model = tf.keras.models.Model(X, Y)
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

#### 막대그래프로 성공확률 확인을 위한 함수

```python
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    plt.imshow(img, cmap=plt.cm.binary)
    
    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'
    
    plt.xlabel('{}{:2.0f}% ({})'.format(labels[predicted_label],
                                   100*np.max(predictions_array),
                                   labels[true_label]),
                                   color=color)
    
def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    thisplot=plt.bar(range(7), predictions_array, color = '#777777')
    plt.ylim([0,0.5])
    predicted_label = np.argmax(predictions_array)
    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python
i=0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions, test_y, test_x)
plt.subplot(1,2,2)
plot_value_array(i, predictions, test_y)
plt.show()
```

```python
num_rows = 21
num_cols =10
num_images = num_rows*num_cols
plt.figure(figsize=(1*3*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plot_image(i, predictions, valid_y, valid_x)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i, predictions, valid_y)
plt.show()
```
